# Compute hamming distance

```python
def hamming_distance(peptide_1, peptide_2):
    return sum([peptide_1[i] != peptide_2[i] for i in range(len(peptide_1))])
```

# Distance from the starting point

### peptides for FYN

```python
distances = []
for peptide in mcmc_hsm_id_fyn_01_peptides:
    distances.append(hamming_distance(mcmc_hsm_id_fyn_01_initial, peptide))
```

```python
plt.scatter(range(len(mcmc_hsm_id_fyn_01_scores)), distances, s=2)
plt.xlabel('iteration')
plt.ylabel('hamming distance')
plt.title(f"Distance from the starting peptide over time \n for peptides targeting FYN with initial peptide: {mcmc_hsm_id_fyn_01_initial}")

plt.show()
# plt.savefig(write_dir / f'plot_{iteration}_vs_rand.png', dpi=300)
```

### peptide sets

```python
peptides_transposed = np.array(mcmc_hsm_id_set_01_peptides).transpose()
```

```python
gene_names = ['FYN', 'SH2D1A','GRB2', 'PIK3R1_nterm', 'PIK3R1_cterm', 'PTPN11_nterm', 'PTPN11_cterm', 'BRDG1', 'SHC1']
```

```python
for target, gene_name in zip(peptides_transposed, gene_names):
    distances = []
    initial = target[0]
    for peptide in target:
        distances.append(hamming_distance(initial, peptide))
    
    plt.scatter(range(len(distances)), distances, s=2)
    plt.xlabel('iteration')
    plt.ylabel('hamming distance')
    plt.title(f"Distance from the starting peptide over time \n for {gene_name} with initial peptide: {initial}")
    plt.show()
```
